#### Bài thực hành 1 trên tập dữ liệu ImageNet với chủ đề phân loại ảnh chó mèo áp dụng mô hình VGG-16. Mục tiêu thực hành là áp dụng mô hình VGG-16 để dự đoán một tấm ảnh chó/mèo.
Đối với ImageNet dataset:
- Tập dữ liệu do đại học Stanford thu thập trên Internet.
- Tập dữ liệu bao gồm 1000 class, 1.200.000 ảnh train, 50.000 ảnh valid, 100.000 ảnh test.

Đối với mô hình VGG-16:
- The model đã dành vị trí 1 và 2 trong chủ đề phân loại ảnh tại cuộc thi ILSVRC 2014.
- Team của trường đại học Oxford, mô hình bao gồm 16 layer fully connected (không tính các layer cho tác vụ tiền xử lý)
- Mô hình đơn giản, dễ train, dễ hiểu và được sử dụng rộng rãi ở nhiều ứng dụng phân loại ảnh sau này.
- Các phiên bản VGG khác có thể kể tới như VGG-11, VGG-13, VGG-19,...

### View and preprocessing image

```python
img_path = "./funny_dog.jpg"
img = Image.open(img_path)
display(img)
```

```python
class BaseTransform():
    def __init__(self,resize, mean,std):
        self.base_transform = transforms.Compose([
            transforms.Resize(resize),
            transforms.CenterCrop(resize),
            transforms.ToTensor(),
            transforms.Normalize(mean,std)
        ])
    def __call__(self,img):
        return self.base_transform(img)
```

```python
resize = 224
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
transform = BaseTransform(resize, mean,std)
img_transform = transform(img) #chuẩn hóa ảnh thành tensor với class BaseTransform
img_transform_view = transforms.ToPILImage()(torch.clamp(img_transform,0,1))#hàm này để giới hạn các giá trị tensor, 
#nếu nhỏ hơn 0 thì trả về 0 và nếu lớn hơn 1 thì trả về 1, còn lại thì giữ nguyên
display(img_transform_view)
```

```python
import json
labels_path = "./imagenet_classes/imagenet_class_index.json"
with open(labels_path, 'r') as f:
    data = json.load(f)

# Print the parsed data
data #dữ liệu các nhãn cùng key của nó
```

```python
class Predictor():
    def __init__(self,classes):
        self.class_pred = classes
    def prediction(self,outp):
        # Get the index of the predicted class
        pred_idx = torch.argmax(outp)
        label_pred = self.class_pred[str(pred_idx.item())]
        return label_pred
```

```python
vgg16 = models.vgg16(pretrained=True) #pretrained = True nghĩa là sử dụng mô hình đã được huấn luyện trước đó
# nếu pretrained = False thì sẽ khởi tạo một network giống hệt với vgg16 thôi nhưng các tham số là ngẫu nhiên vì chưa được train
```

```python
img_transform = img_transform.unsqueeze(0)
vgg16.eval()
with torch.no_grad():
    output = vgg16(img_transform)
predictor = Predictor(data)
res = predictor.prediction(output)
res
```

Như vậy, đoạn code trên đã mô tả toàn bộ quá trình ứng dụng model VGG-16 vào việc đoán một bức ảnh hình chú chó. Bởi vì model này sinh ra trong ngữ cảnh của tập dữ liệu ImageNet liên quan đến các nhãn (labels) về chó mèo nên ta có thể ứng dụng mô hình VGG-16 luôn để tiến hành dự đoán. Tuy nhiên, trong những bài toán thực tế chúng ta lại phải giải quyết việc phân loại có thể là với các loài vật khác. Vậy có cách nào để giải quyết các bài toán đó dựa trên mô hình VGG-16 không, thì xin thưa là có. Tiếp sau chúng ta sẽ tìm hiểu về 2 phương pháp thường được áp dụng trong Deep Learning liên quan đến việc tận dụng những mô hình đã được công bố có tên là 2 phương pháp transfer learning và fine tuning.
